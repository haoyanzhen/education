## 向github中增加ssh密钥

### Realize a ssh key

A ssh key is devided into a public key with `.pub` suffix and a private key.

- Public key: This resides on the server you want to connect to. It's freely distributable and allows the server to verify your identity.
- Private key: This is kept secret on your local machine. It proves that you hold the corresponding public key.

As formal, connecting to a server means copy the public key to a new mechine.

So one ssh key pair consists of two ssh keys. The public key can connect to more than one mechine, while the private key is unique in your own mechine. It means you can create just one key pair for all of ssh connections. It's also natural to create more ssh key pairs for higher security.

### Create a ssh key

Use bash to automatically create a ssh key pair:

```bash
ssh-keygen -t ed25519 -C "your_email@example.com"
```

`-t` means the format of ssh key. You can choose in [ed25519, 4096].
`-C` means the comment email of ssh key. This option is optional.

### Add this ssh key to you os's ssh agent

Start ssh agent first.

```bash
eval "$(ssh-agent -s)"
```

Then modify your `~/.ssh/config` file.

```bash
vi ~/.ssh/config
```

Add the following words to the config file and save it. Note the `UseKeychain` line could not be added if the ssh key is created with an empty passphrase.

```plaintext
Host github.com
  AddKeysToAgent yes
  UseKeychain yes
  IdentityFile ~/.ssh/id_ed25519
```

Add the ssh key pair to your agent. Again if the ssh key if created with an empty passphrase, the option `--apple-use-keychain` is not needed.

```bash
ssh-add --apple-use-keychain ~/.ssh/id_ed25519
```

Check if your agent load the ssh key successfully.

```bash
ssh-add -l
```

### Add the public key in this ssh key pair to your github

Copy all the text into the public key in this ssh key pair. The key file may be named as `id_ed25519.pub`.

Open your github, click the right top icon, and click `Settings`.

Click `SSH and GPG keys`, click `New SSH key`.

Name title. Choose `Authentication Key`.

Paste the text from the public key into the `Key` box.

Click `Add SSH Key`.
