```python
from matplotlib import pyplot as plt
import numpy as np
```

## 基本要素

```python
# Figure
plt.figure(figsize=(5,3), dpi=150)
plt.scatter(np.arange(1000),np.random.rand(1000),s=2)
```

```python
# Axes
ax = plt.axes()
ax.scatter(np.arange(1000),np.random.rand(1000))
```

```python
# Axis
xax = ax.xaxis
yax = ax.yaxis
xax, yax
```

```python
# Artist
```

```python
# axes sub
plt.subplot(221)
plt.plot(np.arange(10),np.arange(10))
plt.subplot(222)
plt.scatter(np.arange(10),np.arange(10))
plt.subplot(223)
plt.scatter(np.arange(10),np.random.rand(10))
plt.subplot(224)
plt.bar(np.arange(10),np.arange(10))
```

```python
# axes sub
fig, axs = plt.subplots(2,3)
for idx, ax in enumerate(axs.reshape(-1)):
    ax.plot(np.arange(10), np.arange(10)**idx)
```

```python
# axes sub add_axes()
fig = plt.figure()
ax1 = fig.add_axes((0,0,0.3,0.5))
ax1.pie(np.arange(10))
ax2 = fig.add_axes((0.5,0.6,0.3,0.3))
ax2.bar(np.arange(10),np.arange(10))
```

```python

# axes sub add_subplot()
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(8,6))
gs = GridSpec(3, 12)
# gs.update(wspace=0.8)
ax00 = fig.add_subplot(gs[0, :4])
ax01 = fig.add_subplot(gs[:2,4:])
ax02 = fig.add_subplot(gs[1:, :4])
ax03 = fig.add_subplot(gs[2, 4:6])
ax04 = fig.add_subplot(gs[2, 6:])
for ax in [ax00,ax01,ax02,ax03,ax04]:
    ax.plot(np.arange(10),np.arange(10))
    ax.plot(np.arange(10),np.arange(10)**2)
    ax.plot(np.arange(10),np.log10(np.arange(10)))
    ax.scatter(np.arange(10),np.arange(10))
ax00.plot(np.arange(10),np.arange(10)**3)
```

## 图像定制

```python
plt.imshow(np.random.rand(10,10), origin='lower')
plt.title("straight line")
plt.xlabel("x", loc='right')
plt.ylabel("y")
plt.xlim(2,8)
plt.colorbar()
plt.grid(True)
plt.xticks(np.arange(10)*2)
plt.yticks(np.arange(5))
plt.text(13, 4, "random")
```
