```python
def conv_mean(array):
    shape = array.shape
    index_list = np.vstack(np.meshgrid(np.arange(shape[0]-2), np.arange(shape[1]-2))).reshape(2,-1).T + 1
    outputarray = np.empty((shape[0]-2)*(shape[1]-2))
    for idx, i in enumerate(index_list):
        outputarray[idx] = np.sum(array[i[0]-1:i[0]+1, i[1]-1:i[1]+1] * 1/9)
    outputarray = outputarray.reshape(np.array(shape)-2)
    return outputarray

conv_mean(t4)
```
