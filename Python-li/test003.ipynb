## numpy

### numpy随机序列

```python
np.random.rand(6,6) * 2 + 0.5
np.random.uniform(0.5, 2.5, (6,6))
```

```python
plt.hist(np.random.randn(10000)*2+5, bins=20, color='b', alpha=0.5)
plt.hist(np.random.normal(loc=5, scale=2, size=(10000)), bins=20, color='r', alpha=0.5)
```

```python
plt.hist(np.random.poisson(100, size=10000), bins=20)
```

```python
plt.hist(np.random.binomial(1000,p=0.1,size=(1000,2)),bins=20)
```

```python
plt.hist(np.random.exponential(scale=5,size=10000),bins=20)
```
