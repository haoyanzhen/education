# matplotlib简单的绘图

```python
from matplotlib import pyplot as plt
import numpy as np
```

## 常用

```python
a1 = np.arange(30)*0.1
b1 = np.sin(a1)
# plt.plot(a1, b1, 'b^-.')  # cyan
plt.plot(a1, b1, color='red', linewidth=0.7)
```

```python
plt.scatter(a1, b1)
```

```python
a2 = np.vstack(np.meshgrid(np.linspace(0,1,100), np.linspace(0,1,100))).reshape(2,-1).T
b2 = a2[:,0]**2 + a2[:,1]**2
plt.scatter(x=a2[:,0],y=a2[:,1],c=b2, cmap='ocean')
plt.colorbar()
```

```python
# 直方图
a3 = np.random.randn(10000)
plt.hist(a3,bins=20)
```

```python
# 柱状图
a4 = np.linspace(-2,2,5)
b4 = np.array([2,4,10,5,3])
plt.bar(a4,b4)
```

```python
# 步进图
plt.step(a4,b4,where='mid')
```

```python
# 饼图
a5 = np.array([3,2,5])
plt.pie(a5, labels=['apple', 'pear', 'peach'])
```
